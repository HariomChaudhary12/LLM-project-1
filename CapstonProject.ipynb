Bag of Words Implementation

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

df = pd.read_csv('Mobile Reviews Sentiment.csv')


vectorizer = CountVectorizer(stop_words='english')


X_bow = vectorizer.fit_transform(df['review_text'].astype(str))


print("Vocabulary Size:", len(vectorizer.get_feature_names_out()))
print("\nSample Vocabulary Words:\n", vectorizer.get_feature_names_out()[:20])


bow_df = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())


bow_df.head()
```

Load Dataset

```python
import pandas as pd

df = pd.read_csv("Mobile Reviews Sentiment.csv")


df = df.rename(columns={"Review": "text", "Sentiment": "label"})

df.head()
```

Convert DataFrame → HuggingFace Dataset

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df)
```

Train/Test Split

```python
dataset = dataset.train_test_split(test_size=0.2)
train_data = dataset["train"]
test_data = dataset["test"]
```

Load BERT Tokenizer

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

Tokenization Function

```python
def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=128)
```

Create DataLoaders

```python
train_loader = DataLoader(train_data, batch_size=8, shuffle=True)
test_loader = DataLoader(test_data, batch_size=8)
```

Load BERT Model for Fine-Tuning

```python
model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
```

Optimizer + Scheduler

```python
optimizer = AdamW(model.parameters(), lr=2e-5)

num_epochs = 2
num_training_steps = num_epochs * len(train_loader)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps
)
```
