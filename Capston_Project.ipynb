```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# If not loaded yet
df = pd.read_csv('Mobile Reviews Sentiment.csv')

# Step 1: Initialize CountVectorizer
vectorizer = CountVectorizer(stop_words='english')

# Step 2: Fit and transform the 'review_text' column
X_bow = vectorizer.fit_transform(df['review_text'].astype(str))

# Step 3: Show vocabulary details
print("Vocabulary Size:", len(vectorizer.get_feature_names_out()))
print("\nSample Vocabulary Words:\n", vectorizer.get_feature_names_out()[:20])

# Step 4: Convert BoW matrix to DataFrame (optional for small samples)
bow_df = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())

# Display first few rows of BoW matrix
bow_df.head()
```

```python
from gensim.models import Word2Vec
import pandas as pd

df = pd.read_csv('Mobile Reviews Sentiment.csv')

# Prepare text
sentences = df['review_text'].astype(str).str.lower().str.split()

# Train Word2Vec
w2v_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)

print("Word2Vec model trained successfully!")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Load dataset (already loaded as df, so skip if not needed)
# df = pd.read_csv('Mobile Reviews Sentiment.csv')

# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Fit & transform the review text
X_tfidf = tfidf.fit_transform(df['review_text'].astype(str))

# Show TF-IDF details
print("TF-IDF vocabulary size:", len(tfidf.get_feature_names_out()))

print("\nSample TF-IDF words:")
print(tfidf.get_feature_names_out()[:20])

# Convert to DataFrame (optional)
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.get_feature_names_out())

tfidf_df.head()
```
